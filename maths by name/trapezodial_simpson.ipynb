```python
def f(x):
    return 1/(1+x**2)

def simpsons13(f,a,b,n):
    h = float(b-a)/n
    
    result = f(a) + f(b)
    
    for i in range(1,n):
        k = a + i * h
        
        if i%2 == 0:
            result = result + 2 *f(k)
        else:
            result = result + 2 *f(k)
    
    result = (h/3)*result
    
    return (result)

simpsons13(f,0,1,4)
```

```python
def f(x):
    return 1/1+x
def trapezodial(f,a,b,n):
    
    h = (b-a)/n
    
    result = f(a) + f(b)
    
    for i in range(1,n):
        k = a + i*h
        result = result + 2 * f(k)
    
    result=(h/2)*result

    return (result)

trapezodial(f,2,10,8)
```
